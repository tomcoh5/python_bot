```python
from binance import ThreadedWebsocketManager
```

```python
def stream_data(msg):
    
    time = pd.to_datetime(msg["E"], unit = "ms")
    price = msg["c"]
    print("Time: {} | Price: {}".format(time,price))
    return price
```

```python
twm = ThreadedWebsocketManager()
```

```python
twm.start_symbol_miniticker_socket(callback = stream_data, symbol = "SANDUSDT")
```

```python
twm.stop()
```
