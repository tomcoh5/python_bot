```python
import pandas as pd 
import matplotlib.pyplot as plt 
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn")
```

```python
msft = close.MSFT.dropna().to_frame().copy()
```

```python
msft
```

```python
msft.rename(columns = {"MSFT":"Price"}, inplace = True)
```

```python
msft
```

```python
msft.shift(periods=1)
```

```python
msft["P_lag1"] = msft.shift(periods=1)
```

```python
msft
```

```python
msft["P_diff"] = msft.Price.sub(msft.P_lag1)
```

```python
msft
```

```python
msft["Returrns"] = msft.Price.pct_change(periods = 1)
msft
```

```python
msft.drop(columns= ["P_lag1"], inplace = True)
```

```python
msft.drop(columns= ["P_diff"], inplace = True)
```

```python
msft.Price.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```
